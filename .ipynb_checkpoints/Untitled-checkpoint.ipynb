**Записали то же самое в формате отношения.**

```python
df.sort_values(by="IMDb-Rating", ascending=False).head()
```

**Отсортировали данные по столбцу IMDb-Rating в порядке убывания.**

```python
df.sort_values(by=["IMDb-Rating", "ReleaseYear"], ascending=[False, True]).head()
```

**Можем отсортировать по двум столбцам. В нашем случае мы сортируем по убывающему рейтингу и возрастающему году выпуска. Т.е. внутри одного рейтинга, год будет возрастать, что видно на "крёстном отце" и "12 разгневанных мужчин". И именно из-за возрастания года "тёмного рыцая" нолана нет во второй таблице, так как он вышел поздно. Поэтому эта таблица отличается от предыдущей.**

```python
df["IMDb-Rating"].mean()
```

**Средняя оценка в этом датасете. Таким образом, мы можем видеть, что в датасете представлены фильмы с довольно высокой средней оценкой.**

```python
df["ReleaseYear"].mean()
```

**Средний год выпуска в этом датасете. То есть средний год выпуска фильма из этого рейтинга - 1991.**

```python
df[df["Director"] == "FrancisFordCoppola"]["IMDb-Rating"].mean()
```

```python
df[df["Director"] == "StevenSpielberg"]["IMDb-Rating"].mean()
```

```python
df[df["Director"] == "FrankDarabont"]["IMDb-Rating"].mean()
```
